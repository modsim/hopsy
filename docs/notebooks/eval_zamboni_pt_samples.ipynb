```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
num_chains = 10

samples = []

for c in range(num_chains):
    samples.append(pd.read_csv('zamboni_samples_' + str(c) + '.csv'))
```

```python
plt.figure()
for i in range(num_chains):

    c = 10
    col = samples[i].columns[c]
    plt.title(col)
    plt.hist(samples[i][col], density=True, alpha=0.245, label=str(i), bins=100)

plt.legend()
plt.show()
```
